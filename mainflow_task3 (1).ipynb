```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("householdtask3.csv")
display(data.head(10))
```

```python
df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# Verify column names
print("Columns in DataFrame:", df.columns)

# Ensure there are no leading/trailing spaces in column names
df.columns = df.columns.str.strip()
```

```python
# Plotting the Bar Chart
plt.figure(figsize=(10, 5))

# Bar chart for total households (tot_hhs) and ownership (own)
plt.bar(df.index, df['tot_hhs'], color='blue', label='Total Households')
plt.bar(df.index, df['own'], color='orange', label='Own', alpha=0.7)

# Customizing the Bar Chart
plt.xlabel('Index')
plt.ylabel('Number of Households')
plt.title('Total Households and Ownership in 2008')
plt.legend()
plt.show()
```

```python
# Plotting the Line Chart
plt.figure(figsize=(10, 5))

# Line chart for income and expenditure
plt.plot(df.index, df['income'], marker='o', linestyle='-', color='blue', label='Income')
plt.plot(df.index, df['expenditure'], marker='o', linestyle='--', color='red', label='Expenditure')

# Customizing the Line Chart
plt.xlabel('Index')
plt.ylabel('Amount ($)')
plt.title('Income and Expenditure in 2008')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the Scatter Plot
plt.figure(figsize=(10, 5))

# Scatter plot for equivalent income (eqv_income) and equivalent expenditure (eqv_exp)
plt.scatter(df.index, df['eqv_income'], color='blue', label='Equivalent Income', s=100)
plt.scatter(df.index, df['eqv_exp'], color='red', label='Equivalent Expenditure', s=100)

# Customizing the Scatter Plot
plt.xlabel('Index')
plt.ylabel('Amount ($)')
plt.title('Equivalent Income and Expenditure in 2008')
plt.legend()
plt.grid(True)
plt.show()
```
